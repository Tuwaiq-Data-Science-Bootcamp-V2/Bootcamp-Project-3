```python
# Old info before changing data types
df.info()
```

```python
# Here I canged the datatypes appropriately as part of the preproccesing phase. 

df['rooms'] = df['rooms'].astype(int)
df['lounges'] = pd.to_numeric(df['lounges'], errors='coerce')
df['bathrooms'] = pd.to_numeric(df['bathrooms'], errors='coerce')
df['propertyAge'] = df['propertyAge'].astype(int)
df['driverRoom'] = df['driverRoom'].astype(int)
df['tent'] = df['tent'].astype(int)
df['patio'] = df['patio'].astype(int)
df['kitchen'] = df['kitchen'].astype(int)
df['outdoorRoom'] = df['outdoorRoom'].astype(int)
df['garage'] = df['garage'].astype(int)
df['duplex'] = df['duplex'].astype(int)
df['space'] = df['space'].astype(int)
df['maidRoom'] = df['maidRoom'].astype(int)
df['elevator'] = df['elevator'].astype(int)
df['furnihsed'] = df['furnihsed'].astype(int)
df['pool'] = df['pool'].astype(int)
df['basement'] = df['basement'].astype(int)
df['apartments'] = pd.to_numeric(df['apartments'], errors='coerce')
```

```python
df.info()
```

```python
# Missing values before cleaning
df.isnull().sum()
```

```python
# Cleaning missing values 

df["lounges"] = df["lounges"].fillna(df["lounges"].median())
df["bathrooms"] = df["bathrooms"].fillna(df["bathrooms"].median())
df["streetWidth"] = df["streetWidth"].fillna(df["streetWidth"].median())
df["price"] = df["price"].fillna(df["price"].median())
df["square price"] = df["square price"].fillna(df["square price"].median())
```

```python
# Data after handling missing values
df.isnull().sum()
```

# Relationship between Villa's age and number of lounges

```python
sns.scatterplot(x='propertyAge', y='lounges', data=df)

plt.title('Relationship Between Property Age and Lounges')
plt.xlabel('propertyAge')
plt.ylabel('lounges')

plt.show()
```

# Relationship between Villa's price and apartments

```python
sns.scatterplot(x='price', y='apartments', data=df)

corr = df['price'].corr(df['apartments'])

plt.title('Relationship between price and apartments')
plt.xlabel('price')
plt.ylabel('apartments')

plt.show()
```

# Top/Bottom 5 Neighbourhoods per number of rooms

```python
neighbourhood_groups = df.groupby('neighbourhood')['rooms'].mean().reset_index()

# sort the neighbourhoods by average number of rooms
neighbourhood_groups = neighbourhood_groups.sort_values(by='rooms', ascending=False)

# select top 5 neighbourhoods
top_5 = neighbourhood_groups.head(5)

# select bottom 5 neighbourhoods
bottom_5 = neighbourhood_groups.tail(5)

# create bar chart for top 5 neighbourhoods
plt.bar(top_5['neighbourhood'], top_5['rooms'])
plt.title('Average Number of Rooms per Neighbourhood (Top 5)')
plt.xlabel('Neighbourhood')
plt.ylabel('Average Number of Rooms')
plt.show()

# create bar chart for bottom 5 neighbourhoods
plt.bar(bottom_5['neighbourhood'], bottom_5['rooms'])
plt.title('Average Number of Rooms per Neighbourhood (Bottom 5)')
plt.xlabel('Neighbourhood')
plt.ylabel('Average Number of Rooms')
plt.show()
```
